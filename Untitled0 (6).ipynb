#Data preprocessing
 * lower case
 * Tokenize
 * removing special characters
 * removing stop words and punctuation
 * stemming

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

nltk.download('punkt')
nltk.download('stopwords')

def transform_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Filter out non-alphanumeric tokens
    tokens = [word for word in tokens if word.isalnum()]

    # Load stopwords and initialize stemmer
    stop_words = set(stopwords.words('english'))
    ps = PorterStemmer()

    # Remove stopwords and punctuation, then stem the words
    filtered_tokens = [ps.stem(word) for word in tokens if word not in stop_words and word not in string.punctuation]

    # Join the processed tokens back into a single string
    transformed_text = " ".join(filtered_tokens)

    return transformed_text
```

```python
df['text_transform']=df['text'].apply(transform_text)
```

#Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv=CountVectorizer()
tfidf=TfidfVectorizer(max_features=3000)
```

```python
x=tfidf.fit_transform(df['text_transform']).toarray()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x.shape
```

```python
y=df['target'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression


models = {
    "Naive Bayes (Gaussian)": GaussianNB(),
    "Naive Bayes (Bernoulli)": BernoulliNB(),
    "Support Vector Machine": SVC(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    print()
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred) * 100
```

```python
# Import the sklearn.metrics module
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score

models_results = {
    "Model": [],
    "Accuracy (%)": [],
    "Precision Score (%)": [],
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred) * 100
    precision = precision_score(y_test, y_pred) * 100
    models_results["Model"].append(name)
    models_results["Accuracy (%)"].append(accuracy)
    models_results["Precision Score (%)"].append(precision)

# Create a DataFrame
models_df = pd.DataFrame(models_results)

# Display the DataFrame
print(models_df.to_string())
```

```python

def predict_spam(text):
    # Preprocess the text
    transformed_text = transform_text(text)

    # Convert the text to a TF-IDF vector
    x = tfidf.transform([transformed_text]).toarray()

    # Scale the vector
    x = scaler.transform(x)

    # Predict the class
    prediction = models["Naive Bayes (Bernoulli)"].predict(x)[0]

    # Return the prediction and the class name
    if prediction == 0:
        return "Ham", prediction
    else:
        return "Spam", prediction

# Example usage
text = "I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times."
prediction, class_name = predict_spam(text)
print(f"Prediction: {prediction}, Class: {class_name}")
```
